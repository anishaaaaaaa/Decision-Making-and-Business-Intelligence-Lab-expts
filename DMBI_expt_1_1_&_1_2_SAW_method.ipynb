# **SAW method**

```
RN = 68
N=93+RN%11
print("No. of Alternatives (N):",N)
M=70+RN%11
print("No. of Attributes (M):",M)
```

```
w=1
wi=w/M
print("Weights are (wi): ",wi)
```

```
lrange = RN//11
print("Start range: ",lrange)
rrange = RN+RN/11
print("End range: ",rrange)
```

```
import numpy as np
import pandas as pd

data = np.random.randint(lrange, rrange+1, size=(N, M))
df = pd.DataFrame(data, columns=[i for i in range(0,M)])
print("Dataset:")
df
```

```
B = M//2
NB = M//2
print("No. of beneficial attributes (B) =",B)
print("No. of non-beneficial attributes (NB) =",NB)
```

```
arr=[]
for i in range(0,M):
  arr.append(i)

Bdata = np.random.randint(0,M,size=B)
for i in range(0, len(Bdata)):
  if Bdata[i] in arr:
    arr.remove(Bdata[i])

print("Beneficial Attributes: ")
print(Bdata)
print()
print("Non-Beneficial Attributes: ",arr)
```

```
for i in arr:
  column = df[i]
  minval = column.min()
  df[i] = minval/df[i]
```

```
for i in Bdata:
  column = df[i]
  maxval = column.max()
  df[i] = df[i]/maxval
```

```
print("Normalized matrix:")
df
```

```
df["sum"] = df.sum(axis=1)
df["Ai"] = df["sum"]*wi
print("Weighted normalized decision matrix : Calculation of Pi")
df
```

```
df['Rank'] = df['Ai'].rank(ascending = 0) 
df
```

```
df = df.set_index('Rank')
df = df.sort_index() 
print("Rank: ")
df 
```

# **Text Book Sum-**

```
af = pd.DataFrame(columns = ["VC", "CF", "PI"])
af["VC"] = [710,900,1630,1720,120,160]
af["CF"] = [400,415,440,235,1150,1750]
af["PI"] = [28,38,59,43,8,19]
print("Book Sum Dataset:")
af
```

```
w1=0.7142
w2=0.1429
w3=0.1429
```

**VC is a beneficial attribute and CF & PI are non-beneficial attributes**

```
vcmax = int(af["VC"].max())
cfmin = int(af["CF"].min())
pimin = int(af["PI"].min())
print("Min-Max values:")
print("Max of VC: ",vcmax)
print("Min of CF: ",cfmin)
print("Min of PI: ",pimin)
```

```
af["VC"] = af["VC"]/vcmax
af["CF"] = cfmin/af["CF"]
af["PI"] = pimin/af["PI"]
print("Normalized matrix:")
af
```

```
af["sum"] = af.sum(axis=1)
af["VC"] = af["VC"]*w1
af["CF"] = af["CF"]*w2
af["PI"] = af["PI"]*w3
af["Ai"] = af.sum(axis=1)
print("Weighted normalized decision matrix : Calculation of Pi")
af
```

```
af['Rank'] = af['Ai'].rank(ascending = 0) 
af
```

```
af = af.set_index('Rank')
af = af.sort_index() 
print("Rank: ")
af 
```
