# **EXPERIMENT NO. 2A**

**Aim:** To implement AHP method using the given details.

```
RN=68
N=10+RN%3
print("No. of Alternatives (N):",N)
M=7+RN%3
print("No. of Attributes (M):",M)
```

```
lrange = 15-RN%16
print("Start range: ",lrange)
rrange = 30+RN%16
print("End range: ",rrange)
```

```
import numpy as np
import pandas as pd

data = np.random.randint(lrange, rrange+1, size=(N, M))
df = pd.DataFrame(data, columns=[i for i in range(0,M)])
print("Decision matrix:")
df
```

```
pairdata = np.random.randint(2, 9, size=(M, M))
pf = pd.DataFrame(pairdata, columns=[i for i in range(0,M)])
pf['Sum'] = pf.sum(axis=0)
print("Pair wise Comparison Dataset:")
pf
```

```
npf=pf
npf = npf/npf["Sum"]
del npf['Sum']
print("Normalised Pair wise comparison matrix:")
npf
```

```
npf["Criteria Weights"] = npf.mean(axis=1)
print("Normalised Pair wise comparison matrix with Criteria weights:")
npf
```

```
pf["Criteria Weights"] = npf["Criteria Weights"]
cpf = pf.mul(pf["Criteria Weights"], axis=1)
del cpf['Sum']
del cpf['Criteria Weights']
print("Column-wise multiplication with the criteria weights:")
cpf
```

```
cpf["Weighted Sum value"] = cpf.sum(axis=1)
cpf["Criteria Weights"] = npf["Criteria Weights"]
cpf["Ratio"] = cpf["Weighted Sum value"]/cpf['Criteria Weights']
print("Ratio of Weighted Sum value and Criteria Weights:")
cpf
```

```
Ymax=cpf["Ratio"].mean()
print("Ymax =",Ymax)
```

```
CI = Ymax-9/9-1
print("Consistency Index =",CI)
```

```
RI = df.iloc[9][0]
CR = CI/RI
print("Consistency Ratio =",CR)
```

**CR is not less than 0.10, the matrix is not consistent.**
