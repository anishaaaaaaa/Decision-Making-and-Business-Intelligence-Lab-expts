**Aim:** To implement TOPSIS method using AHP (for weights) for a given problem.

```
RN=68
N=10+RN%3
print("No. of Alternatives (N):",N)
M=7+RN%3
print("No. of Attributes (M):",M)
```

```
lrange = 15-RN%16
print("Start range: ",lrange)
rrange = 30+RN%16
print("End range: ",rrange)
```

```
import numpy as np
import pandas as pd

data = np.random.randint(lrange, rrange+1, size=(N, M))
df = pd.DataFrame(data, columns=[i for i in range(0,M)])
print("Decision matrix:")
df
```

```
pairdata = np.random.randint(2, 9, size=(M, M))
pf = pd.DataFrame(pairdata, columns=[i for i in range(0,M)])
print("Pair wise Comparison Dataset:")
pf
```

```
w=1
wi=w/M
print("Weights are (wi): ",wi)
```

```
B = M//2+1
NB = M//2
print("No. of beneficial attributes (B) =",B)
print("No. of non-beneficial attributes (NB) =",NB)
```

```
af = pf**2
af['Sum'] = af.sum(axis=0)
af['(Xij2)^(1/2)'] = af['Sum']**(1/2)
af
```

```
af = af.div(af['(Xij2)^(1/2)'])
del af['(Xij2)^(1/2)']
del af['Sum']
print('Normalized matrix:')
af
```

```
af = af*wi
print("Weighted normalized matrix: ")
af
```

```
arr=[]
for i in range(0,M):
  arr.append(i)

Bdata = np.random.randint(0,M,size=B)
for i in range(0, len(Bdata)):
  if Bdata[i] in arr:
    arr.remove(Bdata[i])

print("Beneficial Attributes: ")
print(Bdata)
print()
print("Non-Beneficial Attributes: ",arr)
```

```
vplus=[]
vminus=[]
for i in range(0,M):
  column = af[i]
  minval = column.min()
  maxval = column.max()
  if i in Bdata:
    vplus.append(maxval)
    vminus.append(minval)
  else:
    vplus.append(minval)
    vminus.append(maxval)
print("Ideal Best and Ideal Worst:")
print("V+ :",vplus)
print("V- :",vminus)
```

```
bf = pd.DataFrame()
bf = af.sub(vplus, axis=1)
bf = bf**2
bf['Sum'] = bf.sum(axis=1)
bf
```

```
cf = pd.DataFrame()
cf = af.sub(vminus, axis=1)
cf = cf**2
cf['Sum'] = cf.sum(axis=1)
cf
```

```
ef = pd.DataFrame(columns=['Si+','Si-','Pi','Rank'])
ef['Si+'] = bf['Sum']**(1/2)
ef['Si-'] = cf['Sum']**(1/2)
ef['Pi'] = ef['Si-']/(ef['Si+']+ef['Si-'])
ef['Rank'] = ef['Pi'].rank(ascending = 0) 
ef
```

**Thus the 7th alternative out of the 9 alternatives has ranked the highest amongst all. Hence, 7th is the best choice.**
